```python
import re
from typing import List, Dict


def chunk_markdown_by_heading(markdown_text: str) -> List[Dict[str, str]]:
    """
    Splits a markdown document into chunks based on level-3 headings (###).

    Returns:
        List of dicts with:
        - chunk_id
        - title
        - content
    """

    # Regex to match headings like: ### **Title**
    heading_pattern = re.compile(r"(###\s+\*\*.*?\*\*)")

    # Split while keeping headings
    parts = heading_pattern.split(markdown_text)

    chunks = []
    chunk_id = 1

    for i in range(1, len(parts), 2):
        title = parts[i].strip()
        content = parts[i + 1].strip()

        chunks.append({
            "chunk_id": chunk_id,
            "title": title.replace("###", "").strip(),
            "content": content
        })
        chunk_id += 1

    return chunks

def chunk_numbered_markdown(markdown_text: str) -> List[Dict[str, str]]:
    """
    Splits markdown into chunks based on numbered bold headings:
    **1. Title**, **3.1. Subtitle**, etc.
    """

    heading_pattern = re.compile(
        r"(\*\*\d+(?:\.\d+)*\.\s+.*?\*\*)"
    )

    parts = heading_pattern.split(markdown_text)

    chunks = []
    chunk_id = 1

    for i in range(1, len(parts), 2):
        title = parts[i].strip()
        content = parts[i + 1].strip()

        chunks.append({
            "chunk_id": chunk_id,
            "title": title.replace("**", "").strip(),
            "content": content
        })
        chunk_id += 1

    return chunks

def chunk_markdown_universal(markdown_text: str) -> List[Dict[str, str]]:
    """
    Universal markdown chunker that supports:
    - ### **Heading**
    - **1. Heading**, **3.1. Heading**
    - **Disclaimer:**, **Acknowledgment:**
    """

    heading_pattern = re.compile(
        r"(###\s+\*\*.*?\*\*"
        r"|\*\*\d+(?:\.\d+)*\.\s+.*?\*\*"
        r"|\*\*[A-Z][^*]{2,}?:\*\*)"
    )

    parts = heading_pattern.split(markdown_text)

    chunks = []
    chunk_id = 1

    for i in range(1, len(parts), 2):
        raw_title = parts[i].strip()
        content = parts[i + 1].strip()

        clean_title = (
            raw_title
            .replace("###", "")
            .replace("**", "")
            .strip()
        )

        chunks.append({
            "chunk_id": chunk_id,
            "title": clean_title,
            "content": content
        })
        chunk_id += 1

    return chunks


file_path = "../generated_data/AI_Tool_Usage_Policy.md"
with open(file_path, "r", encoding="utf-8") as f:
    markdown_text = f.read()

# chunks = chunk_markdown_by_heading(markdown_text)
# if len(chunks) == 0:
#     chunks = chunk_numbered_markdown(markdown_text)
chunks = chunk_markdown_universal(markdown_text)

for chunk in chunks:
    print(f"\n--- Chunk {chunk['chunk_id']} ---")
    print(chunk["title"])
    print(chunk["content"])
```

```python
# get all files with .md extension in a directory
def get_all_files(directory: str, extension: str = ".md") -> List[str]:
    """Returns a list of all files with the given extension in the directory."""
    import os
    return [f for f in os.listdir(directory) if f.endswith(extension)]
```
